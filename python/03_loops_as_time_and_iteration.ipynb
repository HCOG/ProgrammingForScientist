# Lecture 3 — Loops as Time Evolution

**Goal:**  
Use loops to model **time evolution** and **iterative physics processes**.

In physics:
- One equation is rarely enough
- Systems evolve step-by-step
- Computers let us *iterate reality*

## Why Loops Matter in Physics

Many physical problems look like:

$$
x_{n+1} = x_n + \Delta x
$$

or, more generally:

$$
\text{state}_{n+1} = \text{state}_n + \text{change}
$$

This is how we:
- Integrate equations of motion
- Evolve temperature, density, velocity
- Approximate differential equations numerically

Loops are how we implement this idea.

```python
for i in range(1, 6):
    print(i)
```

- `range(5)` generates integers: 0, 1, 2, 3, 4
- Each loop iteration represents **one step**
- In physics, this often means **one timestep**


## Example 1 — Constant Velocity Motion

Physics:
$$
x(t+\Delta t) = x(t) + v\,\Delta t
$$

This is the simplest time-integration scheme.


```python
v = 10.0       # m/s
dt = 0.1       # s
x = 0.0        # m

positions = []

for step in range(100):
    x = x + v * dt
    positions.append(x)

positions[:5], positions[-1]
```

Each loop iteration advances time by `dt`.

This is a **numerical integration** of:
$$
\frac{dx}{dt} = v
$$

Even trivial motion already uses computational thinking.


```python
matrix = [[2,6,12,34,55,66],
 [3,7,13,35,56,67],
 [4,8,14,36,57,68],
 [5,9,15,37,58,69],
 [6,10,16,38,59,70],
 [7,11,17,39,60,71]]

for x in range(6):
    for y in range(6):
        for z  in range(6):
            print(matrix[x][y])
```

## Example 2 — Accelerated Motion (Euler Integration)

Physics:
$$
\frac{dv}{dt} = a, \quad \frac{dx}{dt} = v
$$

Discretised:
$$
v_{n+1} = v_n + a\,\Delta t
$$
$$
x_{n+1} = x_n + v_n\,\Delta t
$$

This is called the **Euler method**.


```python
a = -9.81      # m/s^2
dt = 0.01      # s

x = 0.0
v = 20.0       # upward throw

trajectory = []

for step in range(500):
    v = v + a * dt
    x = x + v * dt
    trajectory.append(x) 

trajectory[:5], trajectory[-1]
```

This code is **not exact physics**.

It is an approximation:
- Smaller `dt` → better accuracy
- Larger `dt` → more numerical error

This tradeoff is central to computational physics.


## Example 3 — Radial Free Fall (Gravity)

Acceleration:
$$
a(r) = -\frac{GM}{r^2}
$$

We update:
- velocity using acceleration
- position using velocity

This is already a **nonlinear problem**.


```python
import math

G = 6.674e-11
M = 5.97e24     # Earth mass (kg)

r = 6.4e6 + 1000   # start 1 km above surface
v = 0.0
dt = 0.01

r_values = []

for step in range(1000):
    a = -G * M / r**2
    v = v + a * dt
    r = r + v * dt
    r_values.append(r)

r_values[:5], r_values[-1]
```

You just simulated:

- a time-dependent force
- a nonlinear differential equation
- numerical instability risk if `dt` is too large

This is the **core of orbit and collapse simulations**.


## `while` Loops — Simulate Until Something Happens

Sometimes we don't know *how long* a system evolves.

Examples:
- until collision
- until temperature drops below a threshold
- until escape occurs

This is when `while` is appropriate.


```python
r = 6.4e6 + 1000
v = 0.0
dt = 0.01

steps = 0

while r > 6.4e6 and steps < 10000:
    a = -G * M / r**2
    v = v + a * dt
    r = r + v * dt
    steps += 1

steps, steps * dt
```

⚠️ Always ensure `while` loops have a **guaranteed exit condition**.

In physics:
- this corresponds to conservation laws
- or physical boundaries


## Example 4 — Iterative Convergence (Physics Solvers)

Many physics problems are solved by iteration:
- equilibrium temperature
- pressure balance
- root-finding

We iterate until **change becomes small**.


```python
T = 1000.0   # K
T_eq = 300.0
alpha = 0.1

history = []

for step in range(200):
    T = T - alpha * (T - T_eq)
    history.append(T)

history, history[-1]
```

```python
T_old = 1000.0   # K
T_NEW = 1000.0
T_eq = 300.0
alpha = 0.1
flag = True
history = []


while flag:
    
    T_old = T_NEW
    T_NEW = T_old- alpha * (T_old - T_eq)
    history.append(T_NEW)
    
    if T_old- T_NEW < 0.01:
        flag = False


history, history[-1]
```

This mirrors:
- relaxation methods
- diffusion processes
- thermal equilibrium
- iterative solvers (Gauss–Seidel later)

Loops are **not programming tricks**.
They are **physical processes encoded in time**.
