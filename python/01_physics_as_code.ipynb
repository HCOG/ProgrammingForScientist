## Why Python in Physics-Based Astrophysics?

Astrophysics is not about programming.
It is about **encoding physical laws** so we can:

- Evaluate them numerically
- Explore parameter space
- Simulate systems that cannot be solved analytically

Python is our *numerical laboratory*.


```python
2 + 3 * 4
```

```python
(2 + 3) * 4
```

```python
import math as m 
m.sqrt(16)
```

Python follows **exactly the same rules as algebra**:

- Order of operations
- Brackets
- Functions (√, sin, cos, …)

This is not programming syntax — this is **mathematics executed**.


```python
G = 6.674e-11      # gravitational constant (m^3 kg^-1 s^-2)
M = 1.989e30      # mass of the Sun (kg)
r = 1.496e11      # distance: 1 AU (m)

F = G * M / r**2
F
```

This line:

    F = G * M / r**2

**is Newtonian gravity written verbatim.**

Python does not "know physics".
It only executes what *you* tell it.


⚠️ Python will happily compute **physically meaningless expressions**.

It is your responsibility to:
- Track units
- Track scales
- Track physical meaning

This discipline is essential in computational physics.


```python
10 / 3
```

```python
0.1 + 0.2 == 0.3
```

Floating-point numbers are **approximations**, just like measurements.

This is NOT a bug — it is reality.

**Never compare real-valued physical quantities using `==`.**
