# Lecture 4 — Arrays as Fields and Phase Space

**Goal:**  
Understand arrays as **discretised physical systems**, not just containers of numbers.

In physics:
- a list → many particles
- an array → a field
- indexing → spatial location or particle identity


## From Single Particle to Many Degrees of Freedom

So far we simulated:
- one particle
- one position
- one velocity

Real physics involves:
- many particles
- fields defined over space
- distributions in phase space

Arrays let us represent this numerically.


````python
import numpy as np
````

## Why Arrays Instead of Lists?

Python lists:
- flexible
- slow for numerical physics

NumPy arrays:
- fixed size
- homogeneous (same type)
- vectorised (operate on all elements at once)

This is why NumPy dominates computational physics.


````python
x_list = [0, 1, 2, 3, 4]
x_array = np.array(x_list)

x_array
````

## Arrays as Discretised Space

Consider a 1D spatial grid:
\[
x_i = x_0 + i\,\Delta x
\]

We approximate continuous space with discrete points.


````python
x = np.linspace(0, 10, 101)  # 0 to 10 m, 101 grid points
dx = x[1] - x[0]

x[:5], dx
````

## Fields on a Grid

A scalar field:
- temperature
- density
- potential

is represented as an array with one value per grid point.


````python
# Temperature field (arbitrary example)
T = 300 + 50 * np.sin(2 * np.pi * x / 10)

T[:5]
````

## Vectorisation = Apply Physics Everywhere

## Phase Space: Position + Velocity

In mechanics, a system state is:
\[
(x, v)
\]

For many particles:
- positions → array
- velocities → array

Together, they define **phase space**.


````python
N = 100

# Random initial positions and velocities
x = np.random.uniform(0, 10, N)
v = np.random.normal(0, 1, N)

x[:5], v[:5]
````

## Evolving Phase Space

Using Euler integration:
\[
x_{n+1} = x_n + v_n \Delta t
\]

All particles update simultaneously using arrays.


````python
dt = 0.01
x = x + v * dt

x[:5]
````

## Forces Acting on Many Particles

Example: harmonic oscillator
\[
F = -kx
\quad\Rightarrow\quad
a = -kx/m
\]

Force and acceleration become array operations.


````python
k = 1.0
m = 1.0

a = -k * x / m
v = v + a * dt
````

## Conditional Indexing = Physics Selection

We often want to apply physics only in certain regions:
- density thresholds
- temperature cuts
- spatial masks

NumPy allows **conditional indexing**.


````python
# Particles with positive position
mask = x > 5

mask[:10]
````

````python
# Apply force only where x > 5
v[mask] *= 0.9
````
